```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import sweetviz as sv
import plotly.io as pio
pio.renderers.default = "browser"
```

```python
weather = pd.read_csv('weather.csv', encoding='unicode-escape')
weather
```

```python
weather.columns
```

```python
monthly_rainfall = px.bar(
    data_frame=weather,
    x = 'Month',
    y = 'Rainfall amount (millimetres)',
    color = 'Year',
    animation_frame='Year',
    hover_data=['Period over which rainfall was measured (days)','Period over which rainfall was measured (days)']
)
monthly_rainfall.show()
```

```python
daily_rainfall = px.bar(
    data_frame=weather,
    x = 'Day',
    y = 'Rainfall amount (millimetres)',
    color = 'Month',
    animation_frame='Year',
    hover_data=['Period over which rainfall was measured (days)','Period over which rainfall was measured (days)']
)
daily_rainfall.update_layout(transition = {'duration': 5000})
daily_rainfall.write_html('dailyu.html')
```

```python
rain_by_year = weather.groupby(by='Year')['Rainfall amount (millimetres)'].sum().reset_index()
rain_by_year_bar = px.bar(
    data_frame=rain_by_year,
    x='Year',
    y='Rainfall amount (millimetres)',
    title = 'Total Rain By Year'
)
rain_by_year_bar.show()
```

```python
rain_by_year_avg = weather.groupby(by='Year')['Rainfall amount (millimetres)'].mean().reset_index()
rain_by_year_avg_line = px.line(
    data_frame=rain_by_year_avg,
    x='Year',
    y='Rainfall amount (millimetres)',
    title= 'Average Rain By Year'
)
rain_by_year_avg_line.write_html('avg-rain-year.html')
```

```python
rain_by_year_month_avg = weather.groupby(['Year', 'Month'])['Rainfall amount (millimetres)'].mean().reset_index()
rain_by_year_month_avg_bar = px.bar(
    data_frame=rain_by_year_month_avg,
    x='Month',
    y='Rainfall amount (millimetres)',
    color = 'Year',
    title = 'Average Rain By Month',
    animation_frame='Year'
)
```
